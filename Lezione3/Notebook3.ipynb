In the two exercises below, the temporal evolution of an asset price is simulated within a given time interval. The objective is to calculate the price of a European *call* or *put* option.

**European Options**
The seller of the option, known as the writer, grants the buyer, known as the holder, the right to buy (call option) or sell (put option) a security at a certain future time $T$ for a predetermined exercise price $K$ (strike price).
At time $T$, the buyer decides whether to sell or buy based on the profit (indicated below) they would obtain ($S(t)$ represents the price of the security at time $t$):
   * Call option: $\max{\{0, S(T)-K\}}$
   * Put option: $\max{\{0, K-S(T)\}}$
   
Over time, an interest rate $r$ is applied (thus, the profit is reduced by a factor of $e^{-rt}$).

Therefore, the problem arises for the holder of determining the price at which to transfer this right on the security, taking into account both interest rates and possible fluctuations in the underlying asset price.

The simplest model to tackle this type of problem is the Black-Scholes model, which assumes a price movement following a geometric Brownian motion.

According to this model, we have:

* Call option price:
    $$C\left[S(t), t\right] = S(t)N(d_1) - Ke^{-r(T-t)}N(d_2)$$

* Put option price:
    $$P\left[S(t), t\right] = S(t)\left[N(d_1) - 1\right] - Ke^{-r(T-t)}\left[N(d_2) - 1\right]$$

where:
$$d_1 = \frac{1}{\sigma\sqrt{T-t}}\left[\ln\left(\frac{S(t)}{K}\right) + \left(r + \frac{\sigma^2}{2}(T-t)\right)\right]$$

and
$$d_2 = d_1 - \sigma\sqrt{T-t}$$

In these formulas, $N(x)$ represents the cumulative distribution function of the standard normal distribution $\mathcal{N}(\mu=0,\sigma^2=1)$, defined as:
$$N(x) = \int_{-\infty}^{x}e^{-\frac{z^2}{2}}dz = \frac{1}{2}\left[1 + \text{erf}\left(\frac{x}{\sqrt{2}}\right)\right]$$

The function $\text{erf}(x)$ denotes the error function, given by:
$$\text{erf}(x) = \frac{2}{\sqrt{\pi}}\int_{0}^{x}e^{-z^2}dz$$

In order to compare analytical and Monte-Carlo predictions,  we assume, just like in the Black-Scholes theory, that the asset price at time t is given by a geometric brownian motion with drift $r$ and volatility $\sigma$ ($GBM(r,\sigma^2)$). The asset price at t=T can be sampled by means of two equivalent methods. 
The first one consists in directly sampling S(T) as:
$$
S(T) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)T+\sigma \mathcal{N}(0,1) \sqrt{T}\right]$$
where $\mathcal{N}(0,1)$ is a Gaussian distribution with $\mu=0$ and $\sigma=1$. 

On the other hand, S(T) could be discretized into $n$ sub-intervals $t_0=0<t_1 < t_2 < â€¦ < t_n=T$ so that:
$$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma \mathcal{N}(0,1)\sqrt{t_{i+1}-t_i}\right]$$
 
The two methods are equivalent:
$$
\begin{align}
S(T) &= \prod_{i=0}^{n-1}{S(t_i)} \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{n}-t_{n-1})+\sigma \mathcal{N}(0,1)\sqrt{t_{n}-t_{n-1}}\right]\\
&= S(0) \prod_{i=0}^{n-1} \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma \mathcal{N}(0,1)\sqrt{t_{i+1}-t_i}\right]\\
&= S(0) \exp\left\{\sum_{i=0}^{n-1} {\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma \mathcal{N}(0,1)\sqrt{t_{i+1}-t_i}\right]}\right\}\\
&= S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)T+\sigma \mathcal{N}(0,1) \sqrt{T}\right]
\end{align}
$$   

Where for the last equality it was used: $\sum_{i=0}^{n-1}{\sigma \mathcal{N}(0,1)\sqrt{t_{i+1}-t_i}} = \sigma \mathcal{N}(0,\sigma^2=n)\sqrt{t_{i+1}-t_i} = \sigma \sqrt(n) \mathcal{N}(0,1) \sqrt{t_{i+1}-t_i} = \sigma \mathcal{N}(0,1) \sqrt{T}$.

Firstly, we show the results for direct sampling of the GBM. The option parameters chosen are 
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$

By subtracting the analytical values (call:  14.975790778311286, put:  5.4595325819072364) from the estimated one, the agreement between Black-Scholes theory and Monte-Carlo simulation is shown.

```python
f, error = np.loadtxt("put_prices.txt", usecols=(0,1), delimiter=' ', unpack='true')
x = np.arange(len(f))
plt.errorbar(x,f - 5.4595325819072364,yerr=error, color= 'blue', ecolor='#33ccff')
plt.title(r"Figure 1: Estimation of the price of a call option, $C[S(0),0]$")
plt.xlabel('#blocks')
plt.ylabel(r'$C[S(0),0]- (C[S(0),0])_{analytical}$')
plt.grid(True)
plt.show()

f, error = np.loadtxt("call_prices.txt", usecols=(0,1), delimiter=' ', unpack='true')
x = np.arange(len(f))
plt.errorbar(x,f - 14.975790778311286,yerr=error, color= 'blue', ecolor='#33ccff')
plt.xlabel('#blocks')
plt.ylabel(r'$P[S(0),0]- (P[S(0),0])_{analytical}$')
plt.title(r"Figure 2: Estimation of the price of a put option, $P[S(0),0]$")
plt.grid(True)
plt.show()
```
